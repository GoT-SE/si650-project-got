```python
import json

with open('episodes.json', 'r') as f:
    episodes = json.load(f)['episodes']
```

```python
killedList = []
for episode in episodes:
    for s in episode['scenes']:
        for c in s['characters']:
            if 'killedBy' in c:
                killedList.append((episode['seasonNum'], episode['episodeNum'], s))
```
